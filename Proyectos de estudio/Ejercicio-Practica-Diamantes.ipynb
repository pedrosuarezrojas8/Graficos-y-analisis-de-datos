### Variables del dataset

* carat: Peso del diamante en quilates. Es una variable numérica continua.
* cut: Calidad del corte del diamante. Es una variable categórica que puede tomar los siguientes valores:
+ Fair
+ Good
+ Very Good
+ Premium
+ Ideal
* color: Color del diamante, clasificado de acuerdo a la escala de letras del D al J, donde D representa el mejor color (incoloro) y J el peor color (ligeramente teñido). Es una variable categórica.
* clarity: Claridad del diamante, medida de acuerdo a la cantidad y visibilidad de inclusiones y defectos. Es una variable categórica con los siguientes posibles valores:
+ FL (Flawless)
+ IF (Internally Flawless)
+ VVS1 (Very Very Slightly Included 1)
+ VVS2 (Very Very Slightly Included 2)
+ VS1 (Very Slightly Included 1)
+ VS2 (Very Slightly Included 2)
+ SI1 (Slightly Included 1)
+ SI2 (Slightly Included 2)
* depth: Profundidad total del diamante como un porcentaje del promedio del diámetro. Es una variable numérica continua.
* table: Tamaño de la mesa del diamante (parte superior plana) como un porcentaje del diámetro del promedio. Es una variable numérica continua.
* price: Precio del diamante en dólares estadounidenses. Es una variable numérica continua.
* x: Longitud del diamante en milímetros. Es una variable numérica continua.
* y: Ancho del diamante en milímetros. Es una variable numérica continua.
* z: Altura del diamante en milímetros. Es una variable numérica continua.

```python
# Crear categorías de precios
bins = [0, 500, 1000, 5000, 10000, 20000]
labels = ['Bajo', 'Medio-Bajo', 'Medio', 'Medio-Alto', 'Alto']
df['price_category'] = pd.cut(df['price'], bins=bins, labels=labels)
```

```python
#Relación entre carat y precios
plt.figure(figsize=(10, 6))
sns.scatterplot(y = 'price', x = 'carat', hue = 'clarity', data = df)
plt.title('Relación entre carat y precios')
plt.show()
```

```python
#Relacion entre precios y cantidad
plt.figure(figsize=(8, 6))
sns.histplot(df['price'], bins = 60, kde= True)
plt.title('Relacion entre precios y cantidad')
plt.xlabel('Precio')
plt.ylabel('Cantidad')
plt.show()
```

```python
#Relacion entre longitud del diamante y su precio
plt.figure(figsize=(8, 6))
sns.scatterplot(x = 'x', y = 'price', data = df)
plt.title('Relacion entre longitud del diamante y su precio')
plt.xlabel('Longitud')
plt.ylabel('Precio')
plt.show()
```

```python
#Relacion entre profundidad del diamante y claridad
plt.figure(figsize=(8, 6))
sns.boxplot(x = 'clarity', y = 'depth', data = df)
plt.title('Relacion entre profundidad del diamante y claridad')
plt.xlabel('Claridad')
plt.ylabel('Profundidad')
plt.show()
```

```python
#Distribución de precio por categoria y su cantidad
plt.figure(figsize=(8, 6))
df['price_category'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribución de precio por categoria y su cantidad')
plt.xlabel('Categoria')
plt.xticks(rotation=0)
plt.ylabel('Cantidad')
plt.show()
```

```python
#Relacion del tamaño de la mesa del diamente por su profundidad
plt.figure(figsize=(8, 6))
sns.scatterplot(x = 'table', y = 'depth', data = df, color = 'g')
plt.title('Relacion entre el tamaño de la mesa del diamente por su profundidad')
plt.xlabel('Tamaño de mesa')
plt.ylabel('Longitud')
plt.show()
```

```python
#Distribución del precio por categoría de corte
plt.figure(figsize=(8, 6))
sns.violinplot(y = 'price', data = df, hue='cut')
plt.title('Distribución del precio por categoría de corte')
plt.xlabel('Categoria de corte')
plt.ylabel('Precio')
plt.legend(loc = 'upper right', fontsize = 8)
plt.show()
```

```python
#Relacion entre carat, profundidad, y color
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x = 'carat', y = 'depth', hue = 'color')
plt.title('Relacion entre carat, profundidad, y color')
plt.xlabel('Carat')
plt.ylabel('Profundidad')
plt.legend(loc = 'upper right', fontsize = 10, title = 'Color')
plt.show()
```

```python
#Distribución de la categoría de precio por calidad de corte
plt.figure(figsize=(8, 6))
ax = sns.histplot(data=df, x='cut', hue = 'price_category', multiple = 'stack')
plt.title('Distribución de la categoría de precio por calidad de corte')
plt.xlabel('Calidad de corte')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
ax.legend(title = 'Categoria de precio')
plt.show()
```
