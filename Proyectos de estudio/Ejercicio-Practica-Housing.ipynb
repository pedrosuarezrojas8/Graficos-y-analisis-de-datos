```python
plt.figure(figsize=(10,8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
plt.title('Mapa de calor de correlaccion entre variables')
plt.show()
```

```python
plt.hist(df['CRIM'], bins = 30)
plt.title('Distribución por tasa de criminalidad')
plt.xlabel('Tasa de criminalidad per capita')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.boxplot(df['NOX'])
plt.title('Boxplot de oxidos nitricos')
plt.ylabel('Concentración')
plt.show()
```

```python
plt.hist(df['AGE'], bins = 20, edgecolor = 'black')
plt.title('Distribución de edad de las viviendas')
plt.xlabel('Proporción de unidades construidas antes de 1940')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.hist(df['RAD'], bins = 20, edgecolor = 'black', color = 'orange')
plt.title('Accesibilidad a carreteras')
plt.xlabel('Indice de accesibilidad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.scatter(df['RAD'], df['MEDV'], alpha = 0.5)
plt.title('Accesibilidad a carreteras versus precio')
plt.xlabel('Indice de accesibilidad a carreteras')
plt.ylabel('Precio medio')
plt.show()
```

```python
plt.scatter(df['RM'], df['MEDV'], alpha = 0.5)
plt.title('Número de habitaciones versus precio medio')
plt.xlabel('Número de habitaciones')
plt.ylabel('Precio medio')
plt.show()
```

```python
plt.scatter(df['DIS'], df['MEDV'], alpha = 0.5, color = 'green')
plt.title('Distancia a centros de empleo versus precio medio')
plt.xlabel('Distancia a centros de empleo')
plt.ylabel('Precio medio')
plt.show()
```

```python
plt.scatter(df['LSAT'], df['MEDV'], alpha=0.5, color='purple')
plt.title('Porcentaje de status inferior versus precio medio')
plt.xlabel('Porcentaje de status ')
plt.ylabel('Precio medio')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x = 'RAD', y= 'MEDV', data = df)
plt.title('Valor medio de viviendas por carretera radial')
plt.xlabel('Indice de acceso a carreteras ')
plt.ylabel('Precio medio')
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.regplot(x = 'CRIM', y= 'MEDV', data = df, scatter_kws={'s': 20}, line_kws={'color':'green'})
plt.title('Indice de criminalidad por valor medio de viviendas')
plt.xlabel('Indice de criminalidad ')
plt.ylabel('Precio medio')
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.regplot(x = 'TAX', y= 'MEDV', data = df, scatter_kws={'s': 20}, line_kws={'color':'red'})
plt.title('Tasa de impuesto a la propiedad por valor medio de viviendas')
plt.xlabel('Tasa de impuesto  ')
plt.ylabel('Precio medio')
plt.show()
```

```python
g = sns.PairGrid(df, y_vars=['MEDV'],x_vars=['LSAT', 'RM', 'AGE', 'NOX', 'DIS'])
g.map(sns.regplot, line_kws= {'color': 'red'})
g.add_legend()
plt.suptitle('Relacion de variables seleccionadas respecto al valor medio de viviendas')
plt.subplots_adjust(top = 0.8)
plt.show()
```

```python
selected_features = ['CRIM', 'RM', 'AGE', 'DIS', 'MEDV']
sns.pairplot(df[selected_features], diag_kind='kde', kind='reg', plot_kws={'line_kws': {'color': 'red'}})
plt.suptitle('Pairplot final de la información')
plt.subplots_adjust(top = 0.9)
plt.show()
```

* A mayor criminalidad menor precio de las viviendas
* Número de habitaciones aumenta el valor de las viviendas
* Las viviendas más antiguas tienen a tener valores bajos.
* Accesibilidad: a mayor distancia de los centro de empleo está relacionada con menores valores de las viviendas.
