```python
#Importamos librerias
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec

#Cargamos el archivo CSV en un DataFrame
data = pd.read_csv(r"C:\Users\LEON SUAREZ\Desktop\VSC\Visualizacion datos\Proyectos de estudio\train_clean.csv")

data
```

#### Features Description 

* PassengerId: Identificador único para cada pasajero.
* Survived: Indicador de si el pasajero sobrevivió (1) o no (0).
* Pclass: Clase del billete del pasajero (1, 2 o 3), representando la clase de viaje.
* Name: Nombre completo del pasajero.
* Sex: Género del pasajero (masculino o femenino).
* Age: Edad del pasajero en años.
* SibSp: Número de hermanos/cónyuges a bordo con el pasajero.
* Parch: Número de padres/hijos a bordo con el pasajero.
* Ticket: Número del billete.
* Fare: Tarifa pagada por el billete.
* Cabin: Número de cabina asignada al pasajero (puede ser NaN si no se asignó).
* Embarked: Puerto de embarque (C = Cherbourg, Q = Queenstown, S = Southampton).
* Family_Size: Tamaño de la familia

```python
data['Survived'].value_counts().plot(kind = 'pie', autopct='%1.1f%%', startangle = 90)  # Grafica de pastel de la columna 'Survived'
plt.title('Proporción de Sobrevivientes')  # Título del gráfico                         # Startangle rota el gráfico para mejor visualización
plt.ylabel('')  # Elimina la etiqueta del eje y
plt.legend(['No Sobrevivió', 'Sobrevivió'], loc='upper right', fontsize=8)  # Leyenda del gráfico
plt.show()  # Muestra el gráfico
```

```python
#Graficamos con matplotlib
data['Pclass'].value_counts().sort_index().plot(kind='bar')  # Gráfico de barras de la columna 'Pclass' usando sort_index para ordenar las clases
plt.title('Numero de pasajeros por clase')  # Título del gráfico
plt.xlabel('Clase')  # Etiqueta del eje x
plt.ylabel('Cantidad')  # Etiqueta del eje y
plt.show()  # Muestra el gráfico
```

```python
#Graficamos con seaborn
sns.countplot(x='Pclass', data=data, palette='deep')  # Gráfico de conteo de la columna 'Pclass' usando seaborn
plt.title('Numero de pasajeros por clase')  # Título del gráfico
plt.xlabel('Clase')  # Etiqueta del eje x
plt.ylabel('Cantidad')  # Etiqueta del eje y
plt.show()  # Muestra el gráfico
```

```python
data['Age'].value_counts().sort_index()  # Estadísticas descriptivas de la columna 'Age'
```

```python
#Graficamos la distribución de edades con Matplotlib
plt.hist(data['Age'], bins = 40, edgecolor = 'k')
plt.title('Distribución de pasajeros por edad')  # Título del gráfico
plt.xlabel('Edad')  # Etiqueta del eje x
plt.ylabel('Frecuencia')  # Etiqueta del eje y
plt.show()  # Muestra el gráfico
```

```python
# Gráfico de sobrevivientes por edad con seaborn
sns.boxplot(x='Survived', y='Age', data=data, palette='deep')  
plt.title('Numero de sobrevivientes por edad')  # Título del gráfico
plt.xlabel('Sobrevivientes')  # Etiqueta del eje x
plt.ylabel('Edad')  # Etiqueta del eje y
plt.show()  # Muestra el gráfico
```

```python
#Graficamos si sobrevivieron por sexo con seaborn
sns.countplot(x='Survived', hue = 'Sex', data= data, palette = 'deep')
plt.title('Numero de sobrevivientes por genero')  # Título del gráfico
plt.xlabel('Sobrevivientes')  # Etiqueta del eje x
plt.ylabel('Cantidad')  # Etiqueta del eje y
plt.show()  # Muestra el gráfico
```

```python
#Graficamos si sobrevivieron por clase con seaborn
sns.countplot(x='Survived', hue = 'Pclass', data= data, palette = 'deep')
plt.title('Numero de sobrevivientes por clase')  # Título del gráfico
plt.xlabel('Sobrevivientes')  # Etiqueta del eje x
plt.ylabel('Cantidad')  # Etiqueta del eje y
plt.legend(title='Clase', loc='upper right', fontsize=8)  # Leyenda del gráfico
plt.show()  # Muestra el gráfico
```

```python
#Graficamos con seaborn la cantidad de sobrevivientes por clase
sns.countplot(x='Pclass', hue ='Survived', data=data, palette='deep')
plt.title('Supervivientes por clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.show()
```

```python
#Graficamos la relación entre Edad y Tarifa pagada con Matplotlib
plt.scatter(data['Age'], data['Fare'], alpha = 0.5)
plt.title('Relación entre Edad y Tarifa pagada')
plt.xlabel('Edad')
plt.ylabel('Tarifa')
plt.show()
```

```python
#Graficamos la relación entre Tarifa pagada y Clase con seaborn
sns.boxplot(x='Pclass', y='Fare', data=data, palette='deep')
plt.title('Tarifas por clase')
plt.xlabel('Clase')
plt.ylabel('Tarifa')
plt.show()
```

```python
# Correlación entre variables numéricas mediante una matriz de correlación y graficamos con un heatmap de seaborn
numeric_df = data.select_dtypes(include=[np.number])  # Selecciona solo las columnas numéricas para el análisis de correlación
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap = 'coolwarm') #annot=True para mostrar los valores de correlación en cada celda
plt.title('Matriz de correlación')                     #fmt=".2f" para formatear los números con dos decimales
plt.show()
```

```python
#Grafico de pasajeros por puerto de embarque con seaborn
sns.countplot(x='Embarked', data=data, palette='deep')
plt.title('Número de pasajeros por puerto de embarque')
plt.xlabel('Puerto de embarque')
plt.ylabel('Cantidad')
plt.show()
```

```python
#Grafico de pasajeros por clase y puerto de embarque con seaborn
sns.catplot(x='Pclass', col ='Embarked', data = data, kind = 'count', palette = 'deep') #count es para gráfico de conteo
plt.suptitle('Número de pasajeros por clase y puerto de embarque', y = 1.05)
plt.show()
```

```python
# Pairplot de variables numéricas agrupadas por supervivencia
sns.pairplot(data, hue='Survived', vars = ['Age', 'Fare', 'Pclass'], palette='deep')
plt.suptitle('Pairplot de variables numéricas por supervivencia', y = 1.05)
plt.show()
```

```python
# Histograma de edades por clase y sexo
g = sns.FacetGrid(data, col='Pclass', row='Sex', margin_titles=True)
g.map(sns.histplot, 'Age', bins=10, color='steelblue')
g.fig.suptitle('Distribución de edades por clase y sexo', y=1.03)  # Título del gráfico
plt.show()
```

```python
# Jointplot de edad y tarifa con densidad KDE
sns.jointplot(x='Age', y='Fare', data=data, kind='kde', palette='deep', fill = True)
plt.suptitle('Distribución conjunta de Edad y Tarifa con KDE')
plt.show()
```

```python
#Gráfico de supervivencia según tamaño de la familia con seaborn
sns.countplot(x='Family_Size', hue='Survived', data=data, palette='deep')
plt.title('Supervivencia según tamaño de la familia')
plt.xlabel('Tamaño de la familia')
plt.ylabel('Cantidad')  
plt.show()
```

```python
#Gráfico de mapa de clústeres de la matriz de correlación
correlation_matrix = numeric_df.corr()
sns.clustermap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', figsize=(6, 6))
plt.title('Mapa de clústeres de la matriz de correlación')
plt.show()
```

```python
#Graficos de distribución de edad en subplots con gridspec
fig = plt.figure(figsize=(20, 6)) #Creamos una figura con un tamaño específico
gs = gridspec.GridSpec(nrows=1, ncols=3, figure=fig) #Usamos gridspec para crear una cuadrícula de subplots y definir su tamaño
#Creamos los subplots
ax1 = fig.add_subplot(gs[0, 0])
sns.distplot(data['Age'])
#Graficamos el segundo subplot
ax2 = fig.add_subplot(gs[0, 1])
sns.boxplot(x='Pclass', y ='Age', hue = 'Sex', data=data, palette='deep')
#Graficamos el tercer subplot
ax3 = fig.add_subplot(gs[0, 2])
sns.swarmplot(x='Pclass', y ='Age', hue = 'Survived', data=data, palette='deep', dodge=True)
plt.legend(title = 'Sobrevivió?', labels = ['No', 'Sí']) 
```

Las personas que tendieron a sobrevivir pertenecian a primera clase, tenian menos edad y mayoritariamente eran mujeres

```python
#Graficamos con violinplot
sns.violinplot(x='Pclass', y = 'Age', hue='Survived', data = data, palette = 'deep')
plt.title('Distribución de edad por clase y supervivencia')
plt.xlabel('Clase')
plt.ylabel('Edad')
plt.show()
```
