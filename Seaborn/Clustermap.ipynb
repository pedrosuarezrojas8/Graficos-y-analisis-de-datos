### Clustermap 
Ejemplo 1: Clustermap de la Matriz de Correlación del Dataset iris

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset 'iris' de Seaborn
iris = sns.load_dataset('iris')

# Seleccionar solo las columnas numéricas
numeric_columns = iris.select_dtypes(include=['float64', 'int64'])

# Calcular la matriz de correlación
correlation_matrix = numeric_columns.corr()

# Crear el clustermap avanzado
sns.clustermap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=.5, linecolor='black', cbar_kws={'label': 'Correlación'})

# Añadir título
plt.title('Clustermap de la Matriz de Correlación del Dataset Iris', y=1.05)

# Mostrar el gráfico
plt.show()
```

### Ejemplo 2
Dataset Flights

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset 'flights' de Seaborn
flights = sns.load_dataset('flights')

# Crear una tabla pivote con los datos de vuelos
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

# Crear el clustermap avanzado con los datos de vuelos
sns.clustermap(flights_pivot, cmap="YlGnBu", linewidths=.5, linecolor='black', cbar_kws={'label': 'Número de Pasajeros'}, annot=True, fmt='d')

# Añadir título
plt.title('Clustermap del Número de Pasajeros por Mes y Año', y=1.05)

# Mostrar el gráfico
plt.show()
```

#### Explicación del Código
##### Carga y Preprocesamiento de Datos:

* En ambos ejemplos, se carga un dataset (iris y flights) usando sns.load_dataset.
* Para el dataset iris, se seleccionan solo las columnas numéricas y se calcula la matriz de correlación.
* Para el dataset flights, se crea una tabla pivote con los meses como filas, los años como columnas y los números de pasajeros como valores.


Creación del Clustermap:

* sns.clustermap se utiliza para crear un mapa de calor con clustering jerárquico.
* cmap define la paleta de colores.
* linewidths y linecolor añaden líneas entre las celdas del heatmap.
* cbar_kws se usa para personalizar la barra de color.
* annot y fmt se utilizan para añadir anotaciones a las celdas del heatmap.


##### Personalización y Visualización:

* Se añaden títulos a los gráficos utilizando plt.title.
* y=1.05 en plt.title ajusta la posición del título para que no se superponga con el gráfico.
