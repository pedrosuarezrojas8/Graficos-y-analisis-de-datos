### Regplot y Lmplot: Visualización de Regresiones Lineales con Dispersión


### Regplot

Ejemplo básico

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset 'tips' de Seaborn
tips = sns.load_dataset('tips')

# Crear un regplot de la propina en función del total de la cuenta
plt.figure(figsize=(8, 6))
sns.regplot(x='total_bill', y='tip', data=tips)

# Añadir título y etiquetas
plt.title('Relación entre Total de la Cuenta y Propina')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Propina')

# Mostrar el gráfico
plt.show()
```

### Regplot

Ejemplo avanzado

```python
# Crear un regplot avanzado con ajuste polinómico
plt.figure(figsize=(8, 6))
sns.regplot(x='total_bill', y='tip', data=tips, order=2, ci=None, scatter_kws={'s': 50}, line_kws={'color': 'red'})

# Añadir título y etiquetas
plt.title('Relación entre Total de la Cuenta y Propina (Regresión Polinómica)')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Propina')

# Mostrar el gráfico
plt.show()
```

### LMplot

Ejemplo básico

```python
# Crear un lmplot de la propina en función del total de la cuenta, separado por sexo
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips, height=6, aspect=1.5)

# Añadir título y etiquetas
plt.title('Relación entre Total de la Cuenta y Propina Separado por Sexo')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Propina')

# Mostrar el gráfico
plt.show()
```

### Regplot

Ejemplo avanzado

```python
# Crear un lmplot avanzado de la propina en función del total de la cuenta, separado por día de la semana y ajuste robusto
sns.lmplot(x='total_bill', y='tip', hue='sex', col='day', data=tips, height=4, aspect=0.8, robust=True, palette='coolwarm', markers=['o', 'x'], scatter_kws={'s': 30, 'alpha': 0.7})

# Añadir título y etiquetas
plt.subplots_adjust(top=0.9)
plt.suptitle('Relación entre Total de la Cuenta y Propina Separado por Sexo y Día de la Semana')

# Mostrar el gráfico
plt.show()
```
