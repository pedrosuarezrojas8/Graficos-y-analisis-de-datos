### Parámetros avanzados


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cambiar el estilo global
sns.set_style("whitegrid")

# Crear un gráfico
df = sns.load_dataset('tips')
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Propina vs Total de la Cuenta")
plt.show()
```

#### Control de las Leyendas y Títulos

a.Títulos y Subtítulos:

Puedes añadir títulos y subtítulos a tus gráficos para proporcionar contexto adicional.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", hue="day", data=df)
plt.title("Propina vs Total de la Cuenta", fontsize=16)
plt.suptitle("Análisis de Propinas por Día", fontsize=12)
plt.show()
```

b. Leyendas:

Puedes personalizar la ubicación y el formato de la leyenda.



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", hue="day", style="day", data=df)
plt.title("Propina vs Total de la Cuenta")
plt.legend(title="Día", loc="upper right", fontsize="small")
plt.show()
```

#### Personalización de Etiquetas de Ejes y Ticks

a. Etiquetas de Ejes:

Puedes ajustar las etiquetas de los ejes para mejorar la legibilidad.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Propina vs Total de la Cuenta")
plt.xlabel("Total de la Cuenta ($)", fontsize=12)
plt.ylabel("Propina ($)", fontsize=12)
plt.show()
```

b. Colores en Gráficos Específicos:

Puedes especificar colores en funciones de Seaborn.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", hue="day", palette="coolwarm", data=df)
plt.title("Propina vs Total de la Cuenta")
plt.show()

```

#### Uso de Facetas para Gráficos Multipanel

a. Facetas con sns.FacetGrid:

Puedes crear paneles múltiples para comparar diferentes subconjuntos de datos.

```python
g = sns.FacetGrid(df, col="day", hue="sex", height=4, aspect=1.5)
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
plt.show()
```

b. Facetas con sns.catplot():

catplot también puede ser usado para crear gráficos con múltiples facetas.

```python
sns.catplot(x="total_bill", y="tip", col="day", hue="sex", data=df, height=4, aspect=1.2)
plt.xticks(range(0, 250, 20), [str(i) for i in range(0, 250, 20)])
plt.show()
```

### Ejemplo 1: Gráfico de Dispersión con Puntos Superpuestos
Problema:

Cuando se tienen muchos puntos en un gráfico de dispersión, algunos puntos pueden superponerse, haciendo que sea difícil distinguir entre ellos.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset 'tips'
df = sns.load_dataset('tips')

# Gráfico de dispersión con puntos superpuestos
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", data=df, hue="day", style="day")
plt.title("Propina vs Total de la Cuenta")
plt.show()

```

#### Soluciones:

Uso de Transparencia (Alpha): Ajusta la transparencia de los puntos para que los puntos superpuestos sean visibles.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", data=df, hue="day", style="day", alpha=0.6)
plt.title("Propina vs Total de la Cuenta con Transparencia")
plt.show()
```

Ajuste de Tamaño de los Puntos: Reduce el tamaño de los puntos para minimizar la superposición.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", data=df, hue="day", style="day", s=20)
plt.title("Propina vs Total de la Cuenta con Puntos Más Pequeños")
plt.show()
```

Probar con otro gráfico:

```python
# Crea un gráfico de dispersión con líneas de regresión para cada día
sns.lmplot(x="total_bill", y="tip", hue="day", data=df)

# Configura el título del gráfico
plt.title("Correlación entre Total de la Cuenta y Propina por Día")

# Muestra el gráfico
plt.show()

```

Separamos en sublots:

```python

# Crea subplots para cada día
g = sns.lmplot(x="total_bill", y="tip", hue="day", data=df, col="day", col_wrap=2, height=3, aspect=1)

# Configura el título del gráfico
g.fig.suptitle("Correlación entre Total de la Cuenta y Propina por Día", y=1.05)

# Muestra el gráfico
plt.show()
```

### Ejemplo 3: Gráfico de Líneas con Información Superpuesta

Problema:

En un gráfico de líneas con múltiples series, las líneas pueden superponerse y ser difíciles de distinguir.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="total_bill", y="tip", hue="day", data=df)
plt.title("Propina vs Total de la Cuenta por Día")
plt.show()
```

Soluciones:

Uso de Diferentes Estilos de Línea: Cambia el estilo de las líneas para diferenciarlas.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="total_bill", y="tip", hue="day", style="day", data=df)
plt.title("Propina vs Total de la Cuenta por Día con Estilos de Línea")
plt.show()
```

Uso de Colores Distintos: Utiliza una paleta de colores diferente para cada serie.


```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="total_bill", y="tip", hue="day", palette="Set1", data=df)
plt.title("Propina vs Total de la Cuenta por Día con Colores Distintos")
plt.show()
```

#### Solución: Dividirlo en subplots

```python
# Crear subplots para cada día
g = sns.FacetGrid(df, col="day", hue="day", height=4, aspect=1)

# Dibujar líneas de regresión en cada subplot
g.map(sns.regplot, "total_bill", "tip", scatter_kws={"s": 50})

# Ajustar el diseño
g.add_legend()
g.set_titles(row_template = '{row_name}', col_template = '{col_name}')
plt.show()
```
