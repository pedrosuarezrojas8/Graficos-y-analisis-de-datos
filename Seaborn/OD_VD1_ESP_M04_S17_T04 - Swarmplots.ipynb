### Código básico de Swarmplot
Vamos a utilizar el dataset tips para crear un swarmplot que muestre la distribución de las propinas (tip) por día de la semana (day).

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset 'tips' de Seaborn
tips = sns.load_dataset('tips')

# Crear un swarmplot de la propina por día de la semana
plt.figure(figsize=(8, 6))
sns.swarmplot(x='day', y='tip', data=tips)

# Añadir título y etiquetas
plt.title('Distribución de Propinas por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Propina')

# Mostrar el gráfico
plt.show()
```

* plt.figure(figsize=(8, 6)): Establece el tamaño de la figura.
* sns.swarmplot(x='day', y='tip', data=tips): Crea el swarmplot con las propinas por día de la semana.

### Ejemplo avanzado 
Vamos a crear un swarmplot más avanzado que incluya separación por subcategorías (sexo), personalización de colores y adición de un violinplot de fondo para una visualización combinada.

```python
# Crear un swarmplot avanzado de la propina por día de la semana, separado por sexo
plt.figure(figsize=(10, 6))

# Añadir un violinplot de fondo
sns.violinplot(x='day', y='tip', data=tips, inner=None, color='lightgray')

# Añadir el swarmplot
sns.swarmplot(x='day', y='tip', hue='sex', data=tips, palette='Set1', dodge=True)

# Añadir título y etiquetas
plt.title('Distribución de Propinas por Día de la Semana Separado por Sexo')
plt.xlabel('Día de la Semana')
plt.ylabel('Propina')

# Eliminar duplicados de leyenda
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles[:2], labels[:2])

# Mostrar el gráfico
plt.show()
```

* inner=None: Elimina los elementos internos del violinplot (cajas y bigotes), dejando solo el contorno.
* color='lightgray': Establece el color del violinplot de fondo.

* hue='sex': Separa los puntos por sexo.
* palette='Set1': Establece una paleta de colores personalizada.
* dodge=True: Desplaza los puntos para evitar superposición entre las categorías de hue.
