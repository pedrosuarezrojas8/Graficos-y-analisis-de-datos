### Visualización univariante
#### Distplot: combinación de histograma y KDE

Importo las librerías necesarias

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Traemos el dataset 'tips' de Seaborn

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

Graficamos el total de la cuenta con KDE (Estimación de Densidad del Núcleo)

```python
# Configurar el estilo de los gráficos
sns.set(style='whitegrid')

# Crear un distplot para la variable 'total_bill'
plt.figure(figsize=(10, 6))  # Tamaño de la figura
sns.histplot(tips['total_bill'], kde=True, bins=20, color='skyblue')

# Añadir título y etiquetas
plt.title('Distribución del Total de la Cuenta con KDE')
plt.xlabel('Total de la Cuenta ($)')
plt.ylabel('Frecuencia')

# Mostrar el gráfico
plt.show()
```
