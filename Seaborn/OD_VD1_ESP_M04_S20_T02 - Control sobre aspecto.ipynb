### Control de aspecto

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset 'tips'
df = sns.load_dataset('tips')
```

#### 1. Configuración del Tamaño del Gráfico

a. Uso de Matplotlib:

Seaborn se basa en Matplotlib para la creación de gráficos, así que podemos utilizar funciones de Matplotlib para ajustar el tamaño del gráfico.

```python
plt.figure(figsize=(10, 6))  # Tamaño del gráfico: 10x6 pulgadas
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Propina vs Total de la Cuenta")
plt.show()

```

b. Configuración Global:

Si deseas ajustar el tamaño de todos los gráficos que generas con Seaborn, puedes usar la función seaborn.set().

```python
sns.set(rc={"figure.figsize": (12, 8)})  # Tamaño global del gráfico: 12x8 pulgadas

sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Propina vs Total de la Cuenta")
plt.show()

```

#### 2. Ajuste de la Forma del Gráfico

a. Gráfico de Barras vs Gráfico de Dispersión:

Puedes cambiar el tipo de gráfico según tus necesidades.

```python
# Gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x="day", y="total_bill", data=df)
plt.title("Total de la Cuenta Promedio por Día")
plt.show()

# Gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", hue="day", data=df)
plt.title("Propina vs Total de la Cuenta por Día")
plt.show()
```

b. Uso del Argumento kind:

Con sns.catplot(), puedes ajustar la forma del gráfico utilizando el argumento kind.

```python
# Gráfico de barras con catplot
sns.catplot(x="day", y="total_bill", kind="bar", data=df)
plt.title("Total de la Cuenta Promedio por Día")
plt.show()

```

#### 4. Control de la Orientación

a. Gráficos de Barras:

Puedes cambiar la orientación de los gráficos de barras de vertical a horizontal.



```python
# Gráfico de barras vertical
plt.figure(figsize=(10, 6))
sns.barplot(x="day", y="total_bill", data=df)
plt.title("Total de la Cuenta Promedio por Día")
plt.show()

# Gráfico de barras horizontal
plt.figure(figsize=(10, 6))
sns.barplot(x="total_bill", y="day", data=df)
plt.title("Total de la Cuenta Promedio por Día")
plt.show()

```

b. Uso del Argumento hue:

El argumento hue te permite añadir una dimensión adicional al gráfico, lo cual puede cambiar la orientación visual.



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", hue="day", style="day", data=df)
plt.title("Propina vs Total de la Cuenta por Día")
plt.show()
```

5. Personalización Adicional
Puedes añadir títulos, etiquetas y leyendas para hacer los gráficos más informativos.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", hue="day", style="day", data=df)
plt.title("Propina vs Total de la Cuenta")
plt.xlabel("Total de la Cuenta ($)")
plt.ylabel("Propina ($)")
plt.legend(title="Día")
plt.show()
```
