### Arquitectura de Seaborn
Importo las librerias necesarias

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Traemos el dataset 'tips' de Seaborn

```python
tips = sns.load_dataset('tips')
```

Imprimo el dataframe para visualizar las columnas disponibles

```python
tips
```

Grafico las propinas por día de la semana

```python

# Configuración de estilo y tamaño de la figura
sns.set(style='whitegrid')  # Estilo de fondo del gráfico

plt.figure(figsize=(10, 6))  # Tamaño de la figura (ancho, alto)

# Crear un gráfico de barras de promedio de propinas por día
sns.barplot(x='day', y='tip', data=tips, palette='Set2')  # 'day' y 'tip' son columnas en 'tips'

# Añadir título y etiquetas de los ejes
plt.title('Promedio de Propinas por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Promedio de Propina')

# Mostrar el gráfico
plt.show()
```
