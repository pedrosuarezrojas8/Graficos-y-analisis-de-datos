### Pairplots
Importamos librería y cargamos el dataset Iris


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset 'iris' de Seaborn
iris = sns.load_dataset('iris')

iris
```

Graficamos pairplots

```python
# Crear un pairplot
sns.pairplot(data=iris, hue='species', diag_kind='hist', markers=["o", "s", "D"])

# Mostrar el gráfico
plt.show()
```

### Tipos de Gráficos para Diagonales y No Diagonales
#### Gráficos en las Diagonales
Histogramas

```python
sns.pairplot(data=iris, hue='species', diag_kind='hist')
```

KDE (Estimación de Densidad del Núcleo):



```python
sns.pairplot(data=iris, hue='species', diag_kind='kde')
```

#### Gráficos en las No Diagonales
Diagrama de Dispersión (Scatter Plot)

```python
sns.pairplot(data=iris, hue='species', kind='scatter')
```

Diagrama de Regresión (Regression Plot)

```python
sns.pairplot(data=iris, hue='species', kind='reg')
```
