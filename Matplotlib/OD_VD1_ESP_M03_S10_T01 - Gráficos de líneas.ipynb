### Creación de Gráficos de Líneas Básicos


```python
import matplotlib.pyplot as plt
import numpy as np
```

Generación de Datos de Ejemplo

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

Creación de un Gráfico de Líneas Simple

```python
plt.plot(x, y)
plt.title('Gráfico de Línea Básico')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

### Personalización de Gráficos de Líneas
Cambiar Colores y Estilos de Línea

```python
plt.plot(x, y, color='green', linestyle='--', linewidth=2)
plt.title('Gráfico de Línea Personalizado')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

Añadir Títulos y Etiquetas

```python
plt.plot(x, y)
plt.title('Gráfico de Línea con Títulos y Etiquetas')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.show()
```

### Gráficos de Líneas Avanzados
Múltiples Líneas en un Solo Gráfico

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Seno', color='blue')
plt.plot(x, y2, label='Coseno', color='green')
plt.title('Gráfico con Múltiples Líneas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show()
```

Añadir Anotaciones y Textos

```python
plt.plot(x, y)
plt.title('Gráfico con Anotaciones')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.annotate('Punto máximo', xy=(np.pi/2, 1), xytext=(np.pi/2+1, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

Uso de Subplots para Gráficos Múltiples

```python
fig, axs = plt.subplots(2)

axs[0].plot(x, y1, label='Seno')
axs[0].set_title('Seno')
axs[0].legend()

axs[1].plot(x, y2, label='Coseno', color='red')
axs[1].set_title('Coseno')
axs[1].legend()

plt.tight_layout()
plt.show()
```

### Ejemplos
Ejemplo Avanzado con Múltiples Líneas


```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.plot(x, y1, label='Seno')
plt.plot(x, y2, label='Coseno')
plt.plot(x, y3, label='Tangente', linestyle='--')

plt.title('Ejemplo Avanzado con Múltiples Líneas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.ylim(-10, 10)  # Limitar el eje y para evitar valores extremos de la tangente
plt.grid(True)
plt.show()
```

Ejemplo con Subplots


```python
fig, axs = plt.subplots(3)

axs[0].plot(x, y1, label='Seno')
axs[0].set_title('Seno')
axs[0].legend()

axs[1].plot(x, y2, label='Coseno', color='red')
axs[1].set_title('Coseno')
axs[1].legend()

axs[2].plot(x, y3, label='Tangente', color='green', linestyle='--')
axs[2].set_title('Tangente')
axs[2].legend()

plt.tight_layout()
plt.show()
```
